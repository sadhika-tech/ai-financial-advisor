```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/transactions_raw.csv")

# Basic shape and types
print(df.shape)
print(df.dtypes)
print(df.head(20))

# Null counts — this tells you where cleaning work is
print("\nMissing values:")
print(df.isnull().sum())
print(f"\nNull %: {df.isnull().mean().mul(100).round(1)}")

# Date range
print(f"\nDate range: {df['date'].min()} → {df['date'].max()}")

# Amount distribution
df['amount'].describe()

# Category breakdown (if column exists)
if 'category' in df.columns:
    print(df['category'].value_counts())

# Quick spend chart
df.groupby('category')['amount'].sum().sort_values().plot(kind='barh', figsize=(8,5))
plt.title("Spend by category (raw)")
plt.tight_layout()
plt.savefig("../data/eda_spend_by_category.png")
plt.show()
```
